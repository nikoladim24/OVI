```python
import pandas as pd

data = pd.read_csv('code/data/roads/A1.csv')

data
```

```python
column_data = data['ЈАЗОЛ НА ПОЧЕТОКОТ'].tolist()

```

```python
import graph

jp = graph.Graph()
```

```python
for column in column_data:
    jp.add_vertex(column)
        
```

```python
jp.vertices()
```

```python
column2_data = data['ЈАЗОЛ НА КРАЈОТ'].tolist()
```

```python
for element in column2_data:
    if element not in jp.vertices():
        jp.add_vertex(element)
for i in range(len(column_data)):
    jp.add_edge((column_data[i],column2_data[i]))
```

```python
jp.edges()
```
